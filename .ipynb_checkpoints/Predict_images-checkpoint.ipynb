```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import cv2

from tensorflow.keras.models import load_model
```

```python
# Load model
predict_model = load_model('./models/final_model.h5')
print("model loaded")
```

```python
# Define function: Image pre-processing and normalization
def image_pre(image_name):
    input_size = 150
    im = cv2.imread(image_name) # Load image
    im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
    im_resize = cv2.resize(im, (input_size, input_size)) * 1.0
    im_resize -= [123.68, 116.779, 103.939]
    img_array = np.array([im_resize]) # Image to array
    return img_array
```

```python
# Show image
image_name = "Cute_5.jpg"
image = cv2.imread(image_name)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.show()
```

```python
# Pre-processing image
img = image_pre(image_name)
# Predict image
result = predict_model.predict(img)

if(result <= 0.5):
    print("Beauty image")
else:
    print("Sexy image")
```
